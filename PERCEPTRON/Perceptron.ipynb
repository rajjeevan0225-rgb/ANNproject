```python
import numpy as np

class Perceptron:
    def __init__(self, lr=0.1, epochs=25):
        self.lr = lr
        self.epochs = epochs

    def fit(self, X, y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for _ in range(self.epochs):
            for i in range(len(X)):
                output = np.dot(X[i], self.weights) + self.bias
                pred = 1 if output >= 0 else 0
                diff = y[i] - pred
                self.weights += self.lr * diff * X[i]
                self.bias += self.lr * diff

    def predict(self, X):
        result = np.dot(X, self.weights) + self.bias
        return np.array([1 if i >= 0 else 0 for i in result])


X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,1,1])

model = Perceptron()
model.fit(X, y)

pred = model.predict(X)
print(pred)
print(np.mean(pred == y))
```
